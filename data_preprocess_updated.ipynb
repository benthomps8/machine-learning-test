# Machine Learning Test Project - Adult Income Classification

Data - https://archive.ics.uci.edu/dataset/2/adult

```python
# Import Statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
income = pd.read_csv('adult.DATA', sep=',', header=None)
```

```python
income
```

```python
# Drop NA Values
income.dropna(inplace=True)
```

```python
# Rename Columns
income.rename(columns={0:'Age', 
                       1:'Work Class', 
                       2:'Weight', 
                       3:'Education Type', 
                       4:'Education Years', 
                       5:'Marital Status', 
                       6:'Occupation', 
                       7:'Familial Relationship', 
                       8:'Race', 
                       9:'Sex', 
                       10:'Capital Gain', 
                       11:'Capital Loss', 
                       12:'Hours Per Week', 
                       13:'Native Country', 
                       14:'Income'}, inplace=True)
```

```python
# Drop Capital Gain and Capital Loss due to mostly 0 values
income.drop(['Capital Gain', 'Capital Loss'], axis=1, inplace=True)
```

```python
# Rename Income Column
income['Income'] = income['Income'].replace({' <=50K': 0, ' >50K': 1})
```

```python
income
```

```python
# Train, Tet, Validation Split (80-10-10)
train_income, remaining_df = train_test_split(income, test_size=0.2) 
test_income, val_income = train_test_split(remaining_df, test_size=0.5)
```

```python
train_income
```

```python
# Save CSVs
test_income.to_csv('cleaned_test_income.csv', index=False)
train_income.to_csv('cleaned_train_income.csv', index=False)
val_income.to_csv('cleaned_validation_income.csv', index=False)
```
